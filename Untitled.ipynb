```python
#for data analysis and wrangling
import numpy as np
import pandas as pd

#for data preprocessing
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import LabelEncoder

#for data visualisation
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#machine learning
```

```python
#importing data as pandas dataframe
data = pd.read_csv('../mushroom_train.csv')
data.head()
```

```python
#seperating out the to be classified data
class_data = data['class']
```

```python
#seperating the data on which the class depends
feature_list = data.columns.values
feature_list = np.delete(feature_list,0)
feature_list
feature_data = data[feature_list]
```

```python
#information about the dataset
feature_data.info()
```

```python
feature_data.describe()
```

```python
#there are only 2 numerical fields rest are categorical
#to check if there are null fields in columns
feature_data.isnull().any()
```

```python
#the data doesnot have any null values , except '?' in the column stalkroot
print(len(feature_data['stalk-root'][feature_data['stalk-root'] == '?']))
value = len(feature_data['stalk-root'][feature_data['stalk-root'] == '?'])/len(feature_data['stalk-root'])
print('% of missing value : ' , value*100) 
```

The column "stalk-root" has 30% missing values , therefore this cloumn can lead to misfitting in our classifier
and should be subjected to imputation or should be dropped

```python
total_e = len(data[data['class'] == 'e'])
total_p  = len(data[data['class'] == 'p'])
for i in feature_list:
    print(i)
    for y in feature_data[i].unique():
        print('edible',y,len(data[(data['class'] == 'e') & (data[i] ==y)])/len(data[data[i] == y]))
        print('poisonous',y,len(data[(data['class'] == 'p') & (data[i] ==y)])/len(data[data[i] == y]))
        print()
        
```

Here we get the probabilty of being edible and poisonous given the mushroom has a particular charecteristic(Bayes Thm.) and we can infer that in most cases there is a clear demarcation and the desicion boundary will lead to almost accurate prediction and we can expect to get a very high accuracy coeficient

```python
#the following histograms show the edibility of mushroom based on their radius
g = sns.FacetGrid(data, col='class')
g.map(plt.hist, 'radius', bins=20)
print('_'*40)
#the following histograms show the edibility of mushroom based on their weight
g = sns.FacetGrid(data, col='class')
g.map(plt.hist, 'weight', bins=20)
```

From the above shown histograms we can infer that for lower value of radius we have a higher probablity of having an edible mushroom (in support with Bayes thm.)

```python
#labelencoding categorial data to fit in a model
le = LabelEncoder()
feature_list = np.delete(feature_list,10)
for i in feature_list:
    le.fit(feature_data[i])
    list(le.classes_)
    feature_data[i] = le.transform(feature_data[i])
le.fit(class_data)
list(le.classes_)
class_data = le.transform(class_data)
```

```python
data_labeled = feature_data.copy()
data_labeled['class'] = class_data
```

```python
print (data['class'][0],data_labeled['class'][0])
```

```python
#we can see that for a class value of 1 the mushroom is inedible and therefore a -ve correaltion will lead to better chance of mushroom being edible
for  i in feature_list:
    print(i," : ",data_labeled['class'].corr(data_labeled[i]))
```

```python
#class v/s ring-type histogram
g = sns.FacetGrid(data_labeled, col='class')
g.map(plt.hist, 'ring-type', bins=20)
#class v/s gill-size histogram
g = sns.FacetGrid(data_labeled, col='class')
g.map(plt.hist, 'gill-size', bins=20)
#class v/s gill-color histogram
g = sns.FacetGrid(data_labeled, col='class')
g.map(plt.hist, 'gill-color', bins=20)
```

Histograms also support are correlation values for a +ve correlation as the respective feature increases there is tendency for the mushroom of being Posionus and vice-versa

```python
#doing crossvalidation of our model on the given data
#dropping the stalk-root because of missing values
del feature_data['stalk-root']
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature_data,class_data, test_size=0.33, random_state=42)
clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train,y_train)
```

```python
pred = clf.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
